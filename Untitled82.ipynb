```python
colors = ['\033[96m', '\033[94m', '\033[92m',
          '\033[95m', '\033[93m', '\033[91m']

def my_print(*args, **kwargs):
    args_length = len(args)
    if args_length > 6:
        print(*args, **kwargs)
    else:
        _all = ""
        for arg, col in zip(args, colors[:args_length]):
            _all += col + str(arg) + " "
        print(_all, **kwargs)
```

```python
def to_float(string) -> float | str:
    if string is np.nan:
        return string
    try:
        string = string.strip().replace(",", ".")
        string = float(string)
    except (ValueError, AttributeError):
        pass
    return string

assert isinstance(to_float("0,603605041"), float), "func written wrong"
assert isinstance(to_float(np.nan), float), "func written wrong"
assert isinstance(to_float("asdasd"), str), "func written wrong"
```

```python
def to_timestamp(value: str | float) -> int | float:
    if value is np.nan:
        return value
    dt = datetime.datetime.strptime(value, "%Y-%m-%d %H:%M:%S")
    timestamp = int(datetime.datetime.timestamp(dt))

    return timestamp

assert isinstance(to_timestamp("2012-10-01 00:00:00"), int), "func written wrong"
assert isinstance(to_timestamp(np.nan), float), "func written wrong"
```

```python
mcap_ranked_dict = {
    'Крупнейшие': 6,
    'Крупные': 5,
    'Средние': 4,
    'Малые': 3,
    'Микро': 2,
    'Нано': 1
}
```

```python
def compare(current: str, last: str):
    if current == last:
        return None
    return mcap_ranked_dict[current] > mcap_ranked_dict[last]

assert compare("Крупнейшие", "Крупные") == True, "func written wrong"
assert compare("Нано", "Микро") == False, "func written wrong"
assert compare("Средние", "Средние") == None, "func written wrong"
```
